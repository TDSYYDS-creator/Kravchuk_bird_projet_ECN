```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate T_m (statistic values from noise samples) and t_exp (observed statistic)
m = 199
alpha = 0.05
k = int(alpha * (m + 1))

# Generate sorted sample statistics (e.g., from a normal distribution for illustration)
np.random.seed(42)
T_m = np.sort(np.random.normal(0, 1, m))  # Example sorted noise statistics
t_exp = 1.5  # Example observed statistic (above the threshold)

# Find the threshold
threshold = T_m[-k]

# Plot the sorted noise statistics and the threshold
plt.figure(figsize=(10, 6))
plt.plot(T_m, label="Sorted Noise Sample Statistics (T_m)", color="blue")
plt.axhline(y=threshold, color="red", linestyle="--", label=f"Rejection Threshold (T_m[-{k}])")
plt.axhline(y=t_exp, color="green", linestyle="--", label="Observed Statistic (t_exp)")

# Highlight rejection region
plt.fill_between(np.arange(len(T_m)), threshold, max(T_m), color="red", alpha=0.2, label="Rejection Region")

# Add labels and legend
plt.title("Visualization of Rejection Threshold and Observed Statistic")
plt.xlabel("Sample Index (Sorted by Statistic Value)")
plt.ylabel("Statistic Value")
plt.legend()
plt.grid(True)
plt.show()
```
